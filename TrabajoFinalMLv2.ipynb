```python
datos = datos.drop('CustomerID', axis=1)
```

```python
print(datos['PreferredLoginDevice'].value_counts())
print()
print(datos['PreferredPaymentMode'].value_counts())
```

```python
datos['PreferredLoginDevice'] = datos['PreferredLoginDevice'].replace('Mobile Phone', 'Phone')
datos['PreferredPaymentMode'] = datos['PreferredPaymentMode'].replace('Cash on Delivery', 'COD')
datos['PreferredPaymentMode'] = datos['PreferredPaymentMode'].replace('CC', 'Credit Card')
```

```python
# Imputación diferenciada según lógica de negocio

# Sin actividad previa o cliente nuevo → 0
datos['Tenure'].fillna(0, inplace=True)
datos['HourSpendOnApp'].fillna(0, inplace=True)
datos['CouponUsed'].fillna(0, inplace=True)
datos['OrderCount'].fillna(0, inplace=True)
datos['DaySinceLastOrder'].fillna(0, inplace=True)
datos['OrderAmountHikeFromlastYear'].fillna(0, inplace=True)

# Dato desconocido (dirección incompleta) → mediana
datos['WarehouseToHome'].fillna(
    datos['WarehouseToHome'].median(), inplace=True)

print("Nulos restantes:", datos.isnull().sum().sum())
```

```python
df_final = pd.get_dummies(datos, drop_first=True)
```

```python
print(f"Columnas: {df_final.shape[1]}")
print()
df_final.columns.tolist()
```

```python
df_final.head()
```

```python
# Separamos variables predictoras (X) y objetivo (y)
X = df_final.drop('Churn', axis=1)
y = df_final['Churn']

print(f"X tiene {X.shape[1]} variables y {X.shape[0]} filas")
print(f"Churn rate: {y.mean()*100:.1f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Entrenamos un RF rápido con TODAS las variables
modelo_borrador = RandomForestClassifier(random_state=42)
modelo_borrador.fit(X, y)

# Extraemos la importancia de cada variable
importancias = pd.DataFrame({
    'Variable': X.columns,
    'Importancia': modelo_borrador.feature_importances_
}).sort_values(by='Importancia', ascending=False)

print(importancias)

# Graficamos el top 10
plt.figure(figsize=(10, 6))
plt.barh(importancias['Variable'][:10][::-1], importancias['Importancia'][:10][::-1])
plt.title("Top 10 Variables más importantes")
plt.xlabel("Nivel de Importancia")
plt.show()
```

```python
# Seleccionamos las top 8 variables
top_8_variables = importancias['Variable'].head(8).tolist()
X_top = X[top_8_variables]

print("Variables seleccionadas:")
for i, var in enumerate(top_8_variables, 1):
    imp = importancias[importancias['Variable']==var]['Importancia'].values[0]
    print(f"  {i}. {var} ({imp:.4f})")
```

```python
from sklearn.model_selection import train_test_split

# Primer split: separamos el Test (10%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X_top, y, test_size=0.10, random_state=42)

# Segundo split: del 90% restante, separamos Validation (20/90 ≈ 22%)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.222, random_state=42)

print(f"Training:   {X_train.shape[0]} filas ({X_train.shape[0]/len(y)*100:.0f}%)")
print(f"Validation: {X_val.shape[0]} filas ({X_val.shape[0]/len(y)*100:.0f}%)")
print(f"Test:       {X_test.shape[0]} filas ({X_test.shape[0]/len(y)*100:.0f}%)")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Escalamos (necesario para Logistic Regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Definimos los 3 modelos
modelos = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, random_state=42, eval_metric='logloss')
}

# Entrenamos y evaluamos cada uno con VALIDATION
resultados = {}

for nombre, modelo in modelos.items():
    if nombre == 'Logistic Regression':
        modelo.fit(X_train_scaled, y_train)
        pred = modelo.predict(X_val_scaled)
        prob = modelo.predict_proba(X_val_scaled)[:, 1]
    else:
        modelo.fit(X_train, y_train)
        pred = modelo.predict(X_val)
        prob = modelo.predict_proba(X_val)[:, 1]
    
    resultados[nombre] = {
        'predicciones': pred,
        'probabilidades': prob,
        'accuracy': accuracy_score(y_val, pred),
        'roc_auc': roc_auc_score(y_val, prob)
    }
    
    print(f"\n{'='*60}")
    print(f" {nombre}")
    print(f"{'='*60}")
    print(f"Accuracy:  {resultados[nombre]['accuracy']:.4f}")
    print(f"ROC-AUC:   {resultados[nombre]['roc_auc']:.4f}")
    print(classification_report(y_val, pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Evaluación final del ganador con Test
X_test_scaled = scaler.transform(X_test)

# Random Forest fue el ganador
modelo_ganador = modelos['Random Forest']
pred_test = modelo_ganador.predict(X_test)
prob_test = modelo_ganador.predict_proba(X_test)[:, 1]

print("="*60)
print(" EVALUACIÓN FINAL - Random Forest (Test Set)")
print("="*60)
print(f"Accuracy:  {accuracy_score(y_test, pred_test):.4f}")
print(f"ROC-AUC:   {roc_auc_score(y_test, prob_test):.4f}")
print()
print(classification_report(y_test, pred_test))

# Matriz de confusión
plt.figure(figsize=(6, 4))
matriz = confusion_matrix(y_test, pred_test)
sns.heatmap(matriz, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Se queda', 'Se va'],
            yticklabels=['Se queda', 'Se va'])
plt.title('Matriz de Confusión - Evaluación Final (Test)')
plt.ylabel('Valor Real')
plt.xlabel('Predicción del Modelo')
plt.show()
```

```python
# Curva ROC comparativa
plt.figure(figsize=(8, 6))

for nombre, res in resultados.items():
    fpr, tpr, _ = roc_curve(y_val, res['probabilidades'])
    plt.plot(fpr, tpr, label=f"{nombre} (AUC={res['roc_auc']:.3f})")

plt.plot([0, 1], [0, 1], 'k--', label='Azar (AUC=0.500)')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC - Comparación de Modelos (Validation)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
